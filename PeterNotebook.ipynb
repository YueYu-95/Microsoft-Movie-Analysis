```python
import zipfile
with zipfile.ZipFile('./zippedData/im.db.zip') as zipObj:
    zipObj.extractall(path='./zippedData/')
con = sqlite3.connect('./zippedData/im.db')
mycur = con.cursor() 
mycur.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name;")
available_table=(mycur.fetchall())
available_table
```

```python
df_movie_ratings = pd.read_sql_query("Select * from movie_ratings", con)
df_movie_ratings['movie_id']= df_movie_ratings['movie_id'].apply(str)
df_movie_ratings['averagerating']= df_movie_ratings['averagerating'].apply(str)
df_movie_ratings['numvotes']= df_movie_ratings['numvotes'].apply(str)
df_movie_ratings.head(10)
```

```python
df_known_for = pd.read_sql_query("Select * from known_for", con)
df_known_for['movie_id']= df_known_for['movie_id'].apply(str)
df_known_for['person_id']= df_known_for['person_id'].apply(str)
df_known_for.head(10)
```

```python
df_movie_basics = pd.read_sql_query("Select * from movie_basics", con)
df_movie_basics['movie_id']= df_movie_basics['movie_id'].apply(str)
df_movie_basics['primary_title']= df_movie_basics['primary_title'].apply(str)
df_movie_basics.dropna(subset=['original_title'])
df_movie_basics.head(10)
```

```python
df_directors = pd.read_sql_query("Select * from directors", con)
```

```python
df_movie_akas = pd.read_sql_query("Select * from movie_akas", con)
```

```python
df_persons = pd.read_sql_query("Select * from persons", con)
```

```python
df_principals = pd.read_sql_query("Select * from principals", con)
```

```python
df_writers = pd.read_sql_query("Select * from writers", con)
```

```python
df_mergedframes = df_movie_basics.append([df_movie_ratings, df_known_for, df_directors, df_movie_akas, df_persons, df_principals, df_writers])
display(df_mergedframes)
```
